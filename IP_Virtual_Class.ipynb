## Importing image and converting to gray

```
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
img = cv2.imread("image.png")
print(img.shape)
# cv2 reads in BGR by default
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## Original Image

```
plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.show()
```

## Gray Image

```
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.show()
```

## Using OpenCV

```
# Using mean and median for smoothening
# kernel size is 5, 5

matplotlib.rcParams['figure.figsize']=(10, 10)

kernel = np.ones((5,5),np.float32)/25
blur = cv2.filter2D(img_gray,-1,kernel)
median = cv2.medianBlur(img_gray, 5)

plt.subplot(131)
plt.imshow(img_gray, cmap="gray")
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.subplot(132)
plt.imshow(blur, cmap="gray")
plt.title('Blurred - Mean')
plt.xticks([])
plt.yticks([])
plt.subplot(133)
plt.imshow(median, cmap="gray")
plt.title('Blurred - Median')
plt.xticks([])
plt.yticks([])
plt.show()
```

# Using Skimage
## Median

```
from skimage.morphology import disk
from skimage.filters.rank import mean, median, maximum, minimum, modal
matplotlib.rcParams['figure.figsize']=(10, 10)
plt.subplot(231)
plt.imshow(median(img_gray, disk(1)), vmin=0, vmax=255, cmap=plt.cm.gray)
plt.title("Median - 1")
plt.xticks([])
plt.yticks([])

plt.subplot(232)
plt.imshow(median(img_gray, disk(5)), vmin=0, vmax=255, cmap=plt.cm.gray)
plt.title("Median - 5")
plt.xticks([])
plt.yticks([])

plt.subplot(233)
plt.imshow(median(img_gray, disk(10)), vmin=0, vmax=255, cmap=plt.cm.gray)
plt.title("Median - 10")
plt.xticks([])
plt.yticks([])

plt.subplot(234)
plt.imshow(median(img_gray, disk(15)), vmin=0, vmax=255, cmap=plt.cm.gray)
plt.title("Median - 15")
plt.xticks([])
plt.yticks([])

img_median = median(img_gray, disk(5))
plt.show()
```

## Image Metrics - PSNR

```
from skimage.metrics import peak_signal_noise_ratio
psnrScoreMean = peak_signal_noise_ratio(img_gray, img_mean)
print("Mean PSNR", psnrScoreMean)
psnrScoreMedian = peak_signal_noise_ratio(img_gray, img_median)
print("Median PSNR", psnrScoreMedian)
psnrScoreMax = peak_signal_noise_ratio(img_gray, img_max)
print("Maximum PSNR", psnrScoreMax)
psnrScoreMin = peak_signal_noise_ratio(img_gray, img_min)
print("Mininum PSNR", psnrScoreMin)
```
